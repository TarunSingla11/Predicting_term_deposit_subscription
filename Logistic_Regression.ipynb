```python
def replace_education(education):
    """
    This function codes the highest education attained.
    """
    if education == 'university.degree':
        return 8
    elif education == 'professional.course':
        return 7
    elif education == 'high.school':
        return 6
    elif education == 'basic.9y':
        return 5
    elif education == 'basic.6y':
        return 4
    elif education =='basic.4y':
        return 3
    elif education == 'unknown':
        return 2
    elif education == 'illiterate':
        return 1

def replace_job(job):
    """
    This function codes the highest job earned.
    """
    if job == 'management':
        return 12
    elif job == 'admin.':
        return 11
    elif job == 'entrepreneur':
        return 10
    elif job == 'technician':
        return 9
    elif job == 'services':
        return 8
    elif job == 'self-employed':
        return 7
    elif job == 'blue-collar':
        return 6
    elif job == 'retired':
        return 5
    elif job == 'housemaid':
        return 4
    elif job == 'unemployed':
        return 3
    elif job == 'unknown':
        return 2
    elif job == 'student':
        return 1

def replace_marital(marital):
    if marital == 'married':
        return 4
    elif marital == 'single':
        return 3
    elif marital == 'divorced':
        return 2
    elif marital == 'unknown':
        return 1

def replace_default(default):
    if default == 'no':
        return 0
    elif default == 'yes':
        return 1
    elif default == 'unknown':
        return 2

def replace_housing(housing):
    if housing == 'no':
        return 0
    elif housing == 'yes':
        return 1
    elif housing == 'unknown':
        return 2

def replace_loan(loan):
    if loan == 'no':
        return 0
    elif loan == 'yes':
        return 1
    elif loan == 'unknown':
        return 2

def replace_contact(contact):
    if contact == 'cellular':
        return 1
    elif contact == 'telephone':
        return 2

def replace_poutcome(poutcome):
    if poutcome == 'failure':
        return 0
    elif poutcome == 'success':
        return 1
    elif poutcome == 'nonexistent':
        return 2

df['education'] = df['education'].apply(lambda x: replace_education(x))
df['job'] = df['job'].apply(lambda x: replace_job(x))
df['marital'] = df['marital'].apply(lambda x: replace_marital(x))
df['default'] = df['default'].apply(lambda x: replace_default(x))
df['housing'] = df['housing'].apply(lambda x: replace_housing(x))
df['loan'] = df['loan'].apply(lambda x: replace_loan(x))
df['contact'] = df['contact'].apply(lambda x: replace_contact(x))
df['poutcome'] = df['poutcome'].apply(lambda x: replace_poutcome(x))
```

```python
import numpy as np
import pandas as pd

# Assuming you have already trained the logistic regression model and have coefficients
# Coefficients (from the logistic regression model)
coefficients = {
    'const': -1.4962,
    'age': 0.0094,
    'job': -0.0212,
    'marital': -0.0702,
    'education': 0.0716,
    'default': -0.2493,
    'contact': -0.1431,
    'month': 0.0953,
    'duration': 0.0045,
    'campaign': -0.0470,
    'pdays': -0.0018,
    'previous': 0.1258,
    'poutcome': 0.3551,
    'euribor3m': -0.6042
}

# Function to calculate logit (z)
def calculate_logit(row):
    # Start with the constant (intercept)
    z = coefficients['const']

    # Add the sum of each feature's value multiplied by its coefficient
    for column in row.index:
        if column in coefficients:
            z += coefficients[column] * row[column]

    return z

# Sigmoid function to get probabilities
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Assuming 'X_train' is already encoded and is a DataFrame
# Calculate probabilities for all rows in X_train
probabilities = X_train.apply(lambda row: sigmoid(calculate_logit(row)), axis=1)

# Add the predicted probabilities to your DataFrame
X_train['predicted_probability'] = probabilities

# If you want to see the probabilities
print(X_train[['predicted_probability']].head())
```
